```python
# Filter students who scored more than 80 marks

# Map the filtered students to a new structure: "Neha scored 88"


students = {
    "Amit": 72,
    "Neha": 88,
    "Ravi": 95,
    "Sara": 64,
    "Kiran": 81
}

result = list(map(lambda x: f"{x[0]} scored {x[1]}", filter(lambda x: x[1] > 80, students.items())))

# Here we are getting the output as list, map needs a function, where we are giving the output format as the function
# next we need to apply the function on the filtered people, so we are giving the filter function as the iterable thing

print(result)
```

```python
import csv

filename = "students.csv"

with open(filename, mode="w", newline="") as file:
    writer = csv.writer(file)

    # Write header
    writer.writerow(["ID", "Name", "Email"])

    # Write 25 student records
    for i in range(1, 26):
        name = f"Student{i}"
        email = f"student{i}@example.com"
        writer.writerow([i, name, email])

print("CSV file created successfully!")
```
