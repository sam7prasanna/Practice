```python
f = open("file.txt","w")
f.write("Opening the file\n")   # In this line, we have given \n to enter the data in next line
f.write("Moving to second line")    # Here we have not given \n, so the next data will be written in the same line
f.write("The next one")
f.close()
```

```python
f = open("file.txt","r")
print(f.read())
f.close()
```

```python
# like string, we can use the same functions here
# earlier we entirely printed the text
# we are going to print the content line by line
f = open("file.txt","r")
for line in f:
    print(line.strip())
# using only the line inside the print means, if there is any space, it also will be printed, using strip means, it will remove the spaces
f.close()
```

```python
# To calculate the no of words in the file

f = open("file.txt","r")
words = f.read().split()
print(len(words))
f.close()
```

```python
with open("file.txt", "a") as f:
    f.write("\nnew line added")

d = open("file.txt", "r")
print(d.read())
```

```python
items = ["chicken","mutton","prawn","fish"]
with open("shipping.txt","w") as q:
    for i in items:
        q.write(i + "\n")
```

```python
b = open("shipping.txt","r")
print(b.read())
```

```python
with open("shipping.txt","r") as q:
    line = q.readline()
    print(len(line))
```
