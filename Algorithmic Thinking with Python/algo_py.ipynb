```python
def binary_search(lst, a):
    low = 0
    high = len(lst) - 1

    while low<=high:
        mid = (low+high)//2
        if lst[mid] == a:
            return mid
        elif lst[mid] < a:
            low = mid + 1
        else:
            high = mid - 1
    return "Value not in the list"

lst = [2, 5, 9, 13, 16, 19]

print(binary_search(lst, 3))
```
